```python
board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]


def win_row(board,c):
    for row in board:
        count=0
        for space in row:
            if space == c:
                count+=1
        if count == len(row):
            return True
    return False

def win_column(board,c):
    columns=len(board[0])
    for col in range(columns):
        count=0
        for row in board:
            if row[col] == c:
                count+=1
        if count == len(row):
            return True
    return False
                
def win_diagonal_1(board,c):
    count=0
    for ind in range(len(board)):
        if board [ind][ind]==c:
            count+=1
    if count == len(board):
        return True
    return False
    
def win_diagonal_2(board,c):
    count=0
    for ind in range(len(board)):
        if board [len(board)-1-ind][ind]==c:
            count+=1
    if count == len(board):
        return True
    return False

def is_winner(board, c):
    if win_row(board,c) or win_column(board,c) or win_diagonal_1(board,c) or win_diagonal_2(board,c):
        return True
    return False

def get_winner(board):
    if is_winner(board,'X'):
        return 'X'
    elif is_winner(board,'O'):
        return 'O'
        
print(get_winner(board1))
```

```python
legs = {
    ("upd", "admu"): {
        "travel_time_mins": 10
    },
    ("admu", "dlsu"): {
        "travel_time_mins": 35
    },
    ("dlsu", "upd"): {
        "travel_time_mins": 55
    },
}

def eta(source, destination):
    minutes = 0
    s_found = False
    d_found = False
    for k, v in legs.items():
        (curr_s, curr_d) = k
        if curr_s == source:
            minutes += v["travel_time_mins"]
            source = curr_d
            s_found = True
        if curr_d == destination and s_found:
            d_found = True
            break
    if not d_found:
        for k, v in legs.items():
            (curr_s, curr_d) = k
            if curr_s == source:
                minutes += v["travel_time_mins"]
            if curr_d == destination:
                d_found = True
                break
    return minutes

source = input("Enter source: ")
destination = input("Enter destination: ")
print("ETA:", eta(source, destination))
```
